```python
import numpy as np
```

```python
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))
```

## pandas cumsum() returns cumilative sum over given axis

```python
ts = ts.cumsum()
```

# Basic Plotting : plot

```python
ts.plot()
```

## plot all columns with plot()

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
```

```python
df = df.cumsum()
```

```python
df.plot();
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
```

```python
df3['A'] = pd.Series(list(range(len(df))))
```

## plot one column vs another

```python
df3.plot(x='A', y='B')
```

# Histograms

```python
df4 = pd.DataFrame({'a': np.random.randn(1000) + 1, 'b': np.random.randn(1000), 'c': np.random.randn(1000) - 1}, columns=['a', 'b', 'c'])
```

```python
df4.plot.hist(alpha=0.5)
```

```python
df4.plot.hist(stacked=True, bins=20)
```

## plot different histogram for each column

```python
df4.diff().hist(color='k', alpha=0.5, bins=20)
```

# Scatter Plot

```python
df = pd.DataFrame(np.random.rand(50, 4), columns=['a', 'b', 'c', 'd'])
```

```python
df.plot.scatter(x='a', y='b');
```

### multiple scatter plot
to define multiple scatter plot use 'ax' atttribute

```python
ax = df.plot.scatter(x='a', y='b', color='DarkBlue', label='Group 1');
ax1 = df.plot.scatter(x='c', y='d', color='DarkGreen', label='Group 2', ax=ax);
df.plot.scatter(x='a', y='d', color='Black', label='Group 3', ax=ax1);
```

### scatter plot with 3 variable
represents third variable with color

```python
df.plot.scatter(x='a', y='b', c='c', s=50);
```

bubble chart represents third variable with bubble size

```python
df.plot.scatter(x='a', y='b', s=df['c']*200);
```

# Hexagonal bin Plots
#### alternative to scatter plot if data is dense

```python
df = pd.DataFrame(np.random.randn(1000, 2), columns=['a', 'b'])
```

```python
df['b'] = df['b'] + np.arange(1000)
```

#### the gridsize controls number of hexagons on x axis default 100

```python
df.plot.hexbin(x='a', y='b', gridsize=25)
```
